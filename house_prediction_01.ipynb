# creating a house price prediction using tensorflow and simple nn architecture

importing all important libraries

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

```python
from tensorflow import keras
from keras.layers import Dense
```

# creating a model

```python
model = keras.Sequential([Dense(units=1,input_shape=[1])])

```

```python
# creating smalll dataset
# for 1 room house cost should be 100k+50k
#for 2 room hosue coust shoutd be cost of 1 room house +50k


xs= np.array([1,2,3,4,5,6,7,9],dtype = 'float32')
ys = np.array([150,200,250,300,350,400,450,550],dtype = 'float32')
```

# setting optimizer and loss using compile()

```python
model.compile(optimizer = 'sgd',loss ='mse')
```

# training the model with our dataset

```python
model.fit(xs,ys,epochs=500)
```

# testing the model

```python
model.predict([8,15])
```
